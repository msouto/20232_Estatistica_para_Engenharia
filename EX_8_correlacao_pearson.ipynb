```python
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Gerar dados de exemplo
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + 1 + 0.1 * np.random.randn(100)

# Calcular correlação de Pearson
correlacao, _ = pearsonr(x, y)

# Plotar gráfico de dispersão
plt.scatter(x, y, label=f'Correlação de Pearson = {correlacao:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersão com Correlação de Pearson')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Definir semente para reprodutibilidade
np.random.seed(42)

# Gerar dados aleatórios para três variáveis
dados = np.random.randn(100, 3)

# Criar DataFrame pandas com os dados
df = pd.DataFrame(dados, columns=['Variável1', 'Variável2', 'Variável3'])

# Calcular a matriz de correlação de Pearson
matriz_correlacao = df.corr()

# Exibir a matriz de correlação
print("Matriz de Correlação de Pearson:")
print(matriz_correlacao)

# Criar um gráfico de matriz de calor usando seaborn
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação de Pearson')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Definir semente para reprodutibilidade
np.random.seed(42)

# Gerar dados em que Variável1 e Variável2 estão fortemente correlacionadas
dados = np.random.randn(100, 3)
dados[:, 2] = 2 * dados[:, 0] + np.random.normal(0, 1, 100)  # Variável3 é linearmente dependente de Variável1

# Criar DataFrame pandas com os dados
df = pd.DataFrame(dados, columns=['Variável1', 'Variável2', 'Variável3'])

# Calcular a matriz de correlação de Pearson
matriz_correlacao = df.corr()

# Exibir a matriz de correlação
print("Matriz de Correlação de Pearson:")
print(matriz_correlacao)

# Criar um gráfico de matriz de calor usando seaborn
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação de Pearson')
plt.show()
```

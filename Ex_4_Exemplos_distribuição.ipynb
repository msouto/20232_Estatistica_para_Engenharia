```python
import numpy as np
import matplotlib.pyplot as plt

def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y

# Gerar dados aleatórios
rs1_time_scn1 = np.random.rand(100)  # Substitua 100 pelo tamanho desejado

# Gerar dados aleatórios normalizados
rs1_time_scn1_norm_samples = np.random.normal(
    np.mean(rs1_time_scn1), np.std(rs1_time_scn1), size=10000)

# Calcular ECDF para dados normalizados
x_theor, y_theor = ecdf(rs1_time_scn1_norm_samples)

# Calcular ECDF para dados coletados
x_obs, y_obs = ecdf(rs1_time_scn1)

# Plotar os resultados
plt.plot(x_theor, y_theor)
plt.plot(x_obs, y_obs)
plt.legend(('Distribuição Normal', 'Dados Coletados'), loc='lower right')
plt.xlabel('Valores')
plt.ylabel('Probabilidade acumulada')
plt.title('ECDF dos Dados Coletados e Distribuição Normal Teórica')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y

# Gerar dados com 5 distribuições não normais
num_series = 5
data_series = [np.random.exponential(scale=1, size=100) for _ in range(num_series)]

# Gerar dados aleatórios normalizados
rs1_time_scn1_norm_samples = np.random.normal(
    np.mean(data_series[0]), np.std(data_series[0]), size=10000)

# Calcular ECDF para dados normalizados
x_theor, y_theor = ecdf(rs1_time_scn1_norm_samples)

# Plotar os resultados
plt.plot(x_theor, y_theor, label='Distribuição Normal de Referência')

# Plotar as séries de dados não normais
for i in range(num_series):
    x_obs, y_obs = ecdf(data_series[i])
    plt.plot(x_obs, y_obs, label=f'Dados Não Normais - Série {i + 1}')

plt.legend(loc='lower right')
plt.xlabel('Valores')
plt.ylabel('Probabilidade acumulada')
plt.title('ECDF dos Dados Coletados e Distribuição Normal de Referência')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definir função ECDF
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n + 1) / n
    return x, y

# Gerar dados com diferentes distribuições não normais
np.random.seed(42)  # Para reprodutibilidade
size = 100

data_exp = np.random.exponential(scale=1, size=size)
data_logistic = np.random.logistic(loc=0, scale=1, size=size)
data_gamma = np.random.gamma(shape=2, scale=1, size=size)
data_chisquare = np.random.chisquare(df=2, size=size)
data_cauchy = np.random.standard_cauchy(size=size)

# Gerar dados aleatórios normalizados
data_norm_samples = np.random.normal(np.mean(data_exp), np.std(data_exp), size=10000)

# Calcular ECDF para dados normalizados
x_theor, y_theor = ecdf(data_norm_samples)

# Calcular ECDF para cada série de dados não normais
x_exp, y_exp = ecdf(data_exp)
x_logistic, y_logistic = ecdf(data_logistic)
x_gamma, y_gamma = ecdf(data_gamma)
x_chisquare, y_chisquare = ecdf(data_chisquare)
x_cauchy, y_cauchy = ecdf(data_cauchy)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(x_theor, y_theor, label='Distribuição Normal', linestyle='--')

# Plotar séries de dados não normais
plt.plot(x_exp, y_exp, label='Exponencial')
plt.plot(x_logistic, y_logistic, label='Logística')
plt.plot(x_gamma, y_gamma, label='Gama')
plt.plot(x_chisquare, y_chisquare, label='Qui-Quadrado')
plt.plot(x_cauchy, y_cauchy, label='Cauchy')

plt.legend(loc='lower right')
plt.xlabel('Valores')
plt.ylabel('Probabilidade acumulada')
plt.title('ECDF de Diferentes Distribuições e Distribuição Normal Teórica')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definir função ECDF
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n + 1) / n
    return x, y

# Gerar dados com diferentes distribuições não normais
np.random.seed(42)  # Para reprodutibilidade
size = 100

data_exp = np.random.exponential(scale=1, size=size)
data_logistic = np.random.logistic(loc=0, scale=1, size=size)
data_gamma = np.random.gamma(shape=2, scale=1, size=size)
data_chisquare = np.random.chisquare(df=2, size=size)
data_uniform = np.random.uniform(low=0, high=1, size=size)

# Gerar dados aleatórios normalizados
data_norm_samples = np.random.normal(np.mean(data_exp), np.std(data_exp), size=10000)

# Calcular ECDF para dados normalizados
x_theor, y_theor = ecdf(data_norm_samples)

# Calcular ECDF para cada série de dados não normais
x_exp, y_exp = ecdf(data_exp)
x_logistic, y_logistic = ecdf(data_logistic)
x_gamma, y_gamma = ecdf(data_gamma)
x_chisquare, y_chisquare = ecdf(data_chisquare)
x_uniform, y_uniform = ecdf(data_uniform)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(x_theor, y_theor, label='Distribuição Normal', linestyle='--')

# Plotar séries de dados não normais
plt.plot(x_exp, y_exp, label='Exponencial')
plt.plot(x_logistic, y_logistic, label='Logística')
plt.plot(x_gamma, y_gamma, label='Gama')
plt.plot(x_chisquare, y_chisquare, label='Qui-Quadrado')
plt.plot(x_uniform, y_uniform, label='Uniforme')

plt.legend(loc='lower right')
plt.xlabel('Valores')
plt.ylabel('Probabilidade acumulada')
plt.title('ECDF de Diferentes Distribuições e Distribuição Normal Teórica')
plt.show()
```

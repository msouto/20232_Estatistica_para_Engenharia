```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Gera dados aleatórios para X e Y
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)

# Cria um modelo de regressão linear
modelo = LinearRegression()

# Treina o modelo
modelo.fit(X, Y)

# Faz previsões para novos dados
X_novo = np.array([[0], [2]])
previsoes = modelo.predict(X_novo)

# Plota os dados e a reta de regressão
plt.scatter(X, Y, color='blue', label='Dados Observados')
plt.plot(X_novo, previsoes, color='red', linewidth=2, label='Reta de Regressão')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regressão Linear Simples')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Gerar dados a partir de uma distribuição normal para uma regressão linear bem-sucedida
np.random.seed(42)
dados_x = np.random.normal(size=100)
dados_y = 3 * dados_x + 2 + np.random.randn(100)

# Realizar regressão linear
modelo = LinearRegression()
modelo.fit(dados_x.reshape(-1, 1), dados_y)

# Plotar resultados
plt.scatter(dados_x, dados_y, label='Dados')
plt.plot(dados_x, modelo.predict(dados_x.reshape(-1, 1)), color='red', linewidth=2, label='Regressão Linear')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regressão Linear Bem-Sucedida (Distribuição Normal)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Gerar dados a partir de uma distribuição exponencial para uma regressão linear mal-sucedida
np.random.seed(42)
dados_x = np.random.uniform(low=0, high=10, size=100)
dados_y = 2 * np.exp(0.5 * dados_x) + np.random.randn(100)

# Realizar regressão linear
modelo = LinearRegression()
modelo.fit(dados_x.reshape(-1, 1), dados_y)

# Plotar resultados
plt.scatter(dados_x, dados_y, label='Dados')
plt.plot(dados_x, modelo.predict(dados_x.reshape(-1, 1)), color='red', linewidth=2, label='Regressão Linear')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regressão Linear Mal-Sucedida (Distribuição Exponencial)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import expon, norm, uniform, gamma, lognorm

# Função para gerar dados aleatórios com diferentes distribuições
def gerar_dados(distribuicao, tamanho=100, coef_angular=3, coef_linear=2):
    np.random.seed(42)
    if distribuicao == 'normal':
        dados_x = np.random.normal(size=tamanho)
    elif distribuicao == 'exponencial':
        dados_x = np.random.exponential(size=tamanho)
    elif distribuicao == 'uniforme':
        dados_x = np.random.uniform(size=tamanho)
    elif distribuicao == 'gamma':
        dados_x = np.random.gamma(2, size=tamanho)  # Parâmetro shape=2, ajuste conforme necessário
    elif distribuicao == 'lognormal':
        dados_x = np.random.lognormal(size=tamanho)
    else:
        raise ValueError('Distribuição não suportada')

    dados_y = coef_angular * dados_x + coef_linear + np.random.randn(tamanho)
    return dados_x, dados_y

# Função para realizar regressão linear e plotar resultados
def regressao_linear_e_plot(dados_x, dados_y, distribuicao):
    modelo = LinearRegression()
    modelo.fit(dados_x.reshape(-1, 1), dados_y)

    plt.scatter(dados_x, dados_y, label=f'Dados ({distribuicao})')
    plt.plot(dados_x, modelo.predict(dados_x.reshape(-1, 1)), color='red', linewidth=2, label='Regressão Linear')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Regressão Linear com Distribuição {distribuicao}')
    plt.legend()
    plt.show()

# Gerar dados com diferentes distribuições
dados_x_normal, dados_y_normal = gerar_dados('normal')
dados_x_exponencial, dados_y_exponencial = gerar_dados('exponencial')
dados_x_uniforme, dados_y_uniforme = gerar_dados('uniforme')
dados_x_gamma, dados_y_gamma = gerar_dados('gamma')
dados_x_lognormal, dados_y_lognormal = gerar_dados('lognormal')

# Realizar regressão linear e plotar resultados para cada distribuição
regressao_linear_e_plot(dados_x_normal, dados_y_normal, 'Normal')
regressao_linear_e_plot(dados_x_exponencial, dados_y_exponencial, 'Exponencial')
regressao_linear_e_plot(dados_x_uniforme, dados_y_uniforme, 'Uniforme')
regressao_linear_e_plot(dados_x_gamma, dados_y_gamma, 'Gamma')
regressao_linear_e_plot(dados_x_lognormal, dados_y_lognormal, 'Log-Normal')
```
